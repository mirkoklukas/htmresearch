```python
import math
import json
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import sys
sys.path.append("../")

from unique_hypercube_size import *

%load_ext autoreload
%autoreload 2
```

```python
"""

    Set up the parameters
    
"""

num_trials = 10

phase_resolution = 0.3
phrs = [0.4,0.35,0.3, 0.25, 0.2, 0.15, 0.1, 0.05]


m_max = 5
k_max = 4

ms     = range(4, m_max + 1)
ks     = range(2, k_max + 1)
scales = np.sqrt(2.)**np.arange(m_max)

ms= [5]
# scales = 1. + np.random.sample(m_max)*0.5 - 0.25

print phrs
print scales
```

```python
"""

    Run the experiments and 
    collect the data:
    
        - width 
        - det(R)  
        - volume of embedded unit cube
    
"""

emb_vol = np.zeros((len(phrs), m_max, k_max, num_trials))
width   = np.zeros((len(phrs), m_max, k_max, num_trials))
det_R   = np.zeros((len(phrs), m_max, k_max, num_trials))




for t in xrange(num_trials):
    A = create_random_A(len(scales), max(ks), scales)
    for p, phr in enumerate(phrs):
        A, results = doRandomModuleExperiment(ms, ks, scales, phr, action_tensor=A)

        for (m,k), r in results.iteritems():
            width[p, m-1,k-1, t] = r

        for m in ms:
            for k in ks:
                if 2*m >= k:
                    Q, R   = np.linalg.qr(A[:m,:,:k].reshape((2*m,k)))

                    det_R[p, m-1,k-1,t] = np.absolute(np.linalg.det(R))

                    vol_k     = det_R[p, m-1,k-1,t]
                    vol_fibre = phr**(2*m - k)

                    emb_vol[p, m-1,k-1,t] = vol_k * vol_fibre

    

        


```

```python
"""

    Compute the quantities derived
    from data
    
"""



mean_width   = np.zeros((len(phrs), m_max, k_max))
vol          = np.zeros((len(phrs), m_max, k_max, num_trials))
mean_vol     = np.zeros((len(phrs), m_max, k_max))
mean_emb_vol = np.zeros((len(phrs), m_max, k_max))
mean_det_R   = np.zeros((len(phrs), m_max, k_max))

for p, phr in enumerate(phrs):
    for m in ms:
        for k in ks:
            m_ = m - 1
            k_ = k - 1
            vol[p,m_,k_]          = width[p,m_,k_]**k
            mean_width[p,m_,k_]   = np.mean(width[p,m_,k_])
            mean_vol[p,m_,k_]     = np.mean(vol[p,m_,k_])
            mean_emb_vol[p,m_,k_] = np.mean(emb_vol[p,m_,k_])
            mean_det_R[p,m_,k_]   = np.mean(det_R[p,m_,k_])

        
```

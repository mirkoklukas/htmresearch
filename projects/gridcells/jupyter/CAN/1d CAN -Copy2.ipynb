```python
np.seterr(invalid='raise')


T = 10000

S = np.zeros((T,n))

# S[0] = np.zeros(n)
S[0] = np.random.sample(n)

Beta = np.zeros((T,n))


for t in range(1,T):
    
    s = S[t-1]
    b = np.zeros(n)
    
    tonic = .1
    beta  = Beta[t-1]


    
    s_   = evolve_I(W, b  + tonic, s, A, beta_mask*beta)  
    S[t] = s_


    alpha   = 0.01
    try:
        alpha*(s_ - np.mean(s_))
    except FloatingPointError:
        pass

    Beta[t] = (1. - alpha)*Beta[t-1] - alpha*(s_ - np.mean(s_))



plt.figure(figsize=(20,5))
plt.matshow(S[-2000:].T, cmap="jet")
plt.show()



```

```python




def compute_scales(W):
    gc = W.shape[1]
    scales = np.zeros(gc)
    for i in range(gc):
        f = (W[:,i] > 0.5).astype(float)
        df = f[1:] - f[:-1]
        ind = np.where(df == -1.)[0]
        ind = ind.astype(float)
        if len(ind) > 1:
            scales[i] = np.mean(ind[1:] - ind[:-1])
        else:
            scales[i] = -100

    return scales


plt.style.use('ggplot')

# S_ = S
S_ = S/np.amax(S, axis=1, keepdims=True)


plt.figure(figsize=(20,5))
patch = S_[6000:10000]
plt.imshow(patch, cmap="jet")
plt.show()


scales = compute_scales(patch.T)

# np.clip(scales, -1,1000,scales)

# plt.scatter(np.arange(n), scales + np.random.sample(n)*3., edgecolor="white")
# plt.show()
plt.hist(scales, bins=100);
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


n=20
N=n**2
diam=2.
X = np.mgrid[0:diam:diam/n, 0:diam:diam/n].reshape(2,-1).T


T = np.zeros((N,N))

for i in np.arange(N):
    for j in np.arange(N):
        d = np.linalg.norm(X[i] - X[j])
        T[i,j] = np.exp(-d**2/0.05)


T= (T + T.T)/2.
T = T/np.sum(T, axis=1, keepdims=True)

lam, vec= np.linalg.eigh(T)

plt.plot(lam)
```

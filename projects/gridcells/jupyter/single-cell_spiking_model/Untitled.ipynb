```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


x = np.linspace(0,10,num=100)
tau_s = 1.
tau_l = 2.
mu = 1.5

K_s = 1/tau_s*np.exp(-x/tau_s) 
K_l = 1/tau_l*np.exp(-x/tau_l) 
K   = K_s - mu*K_l

plt.plot(x,K_s, label="Short time filter $K_S$")
plt.plot(x,K_l, label="Long time filter $K_L$")

plt.legend()
plt.show()



plt.plot(K)
plt.plot([0,len(K)],[0,0],'r--',lw=2)
plt.show()
```

```python
duration = 1
dt = 0.001
n  = 100
T  = int(duration/dt)
sigma_loc = 0.1

x_cell  = np.linspace(0.,1.,num=n)
x_agent = np.linspace(0.,1.,num=T)

y_R = np.zeros((T,n))
y_I = np.zeros((T,n))

for t in range(T):
    pr = np.exp( - (x_cell - x_agent[t])**2/(2*sigma_loc**2))
    rnd = np.random.sample(n)
    y_R[t] = (rnd < pr)
    rnd = np.random.sample(n)
    y_I[t] = (rnd < 0.6*pr)

    
# plt.plot(x,pr)

plt.figure(figsize=(20,10))
plt.imshow(y_R.T + y_I.T)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


x = np.linspace(0,10,num=100)
tau_s = 1.
tau_l = 2.
mu = 1.5

K_s = 1/tau_s*np.exp(-x/tau_s) 
K_l = 1/tau_l*np.exp(-x/tau_l) 
K   = K_s - mu*K_l

plt.plot(x,K_s, label="Short time filter $K_S$")
plt.plot(x,K_l, label="Long time filter $K_L$")

plt.legend()
plt.show()



plt.plot(K)
plt.plot([0,len(K)],[0,0],'r--',lw=2)
plt.show()
```

```python
"""Create an environment."""

num_features = 4
feature_dist = 10.
feature_prob = 0.8
env = -1*np.ones((200,200))
colors = [0,1,2,3,4,5,6]

for x in range(env.shape[0]):
    for y in range(env.shape[1]):
        if x%feature_dist == 0 and y%feature_dist == 0 and np.random.sample() < feature_prob:
            f = np.random.randint(num_features)
            env[x,y] = f

# plt.figure(figsize=(15,15))
# plt.imshow(env, cmap="binary")
# plt.show()

plt.figure(figsize=(15,15))
for f in range(num_features):
    xs = np.where(env==f)[0], 
    ys = np.where(env==f)[1]
    plt.scatter(xs,ys, s=10, marker="s",  edgecolor='red',  lw = 0)
# plt.scatter([100],[100], s=500, marker="X", c="red", edgecolor='red',  lw = 0)










```
